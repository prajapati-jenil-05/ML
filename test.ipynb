# **1. Count plot (used for categorical data)**  

Count plot are used when you have categorical variables, and you want to visualize the distribution of these categories.


```python
# Count plot
clean_dataset['source_city'].unique()
```

```python
clean_dataset.nunique()
```

```python
sns.countplot(x='source_city',data=clean_dataset)
```

```python
sns.countplot(x='class',data=clean_dataset)
```

```python
sns.countplot(y='source_city',data=clean_dataset)
```

# 2. **Pie Chart (Categorical data)**   
A pie chart represent the parts of a whole entity

```python
clean_dataset['class'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
clean_dataset['class'].value_counts().plot(kind='line')
```

```python
clean_dataset['source_city'].value_counts().plot(kind='pie',autopct="%.2f")
```

# 3. **Donut plots (used for categorical data)**  
Donut plots are used to show the proportion of each category 

```python
plt.pie(clean_dataset['source_city'].value_counts(),labels=clean_dataset['source_city'].unique(),autopct="%.2f",wedgeprops=dict(width=0.2))
```

# 4. **Histogram (used for numerical data) (Countinuos Variables)**
